```python
from sympy import*
t = symbols("t")
x = Function("x")
m = 1
k = 0.2
k2 = 0.1
```

```python
de = Eq(diff(x(t),t,t),-k*x(t)-k2*diff(x(t),t))
de
```

```python
s = dsolve(de,x(t), ics = {x(0):1, diff(x(t),t).subs(t,0):0.01})
s
```

```python
pt = plot(s.rhs, (t,0,80), show = False)
p1 = plot(exp(-0.05*t), (t,0,80), line_color="red", show =False)
p2 = plot(-exp(-0.05*t), (t,0,80), line_color="red", show =False)
pt.extend(p1)
pt.extend(p2)
pt.show()
```
