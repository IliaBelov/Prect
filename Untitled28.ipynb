```python
# SI
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
 
N = 1000 # Total population, N.
days = 364
 
I0 = 1 # Initial number of infected individuals
S0 = N - I0
 
beta = 0.1 # Contact rate beta
 
t = np.linspace(0, days, days*4)  # A grid of time points (in days)
 
# The SI model differential equations.
def deriv(y, t, N, beta):
    S, I = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N
    return dSdt, dIdt
 
y0 = S0, I0 # Initial conditions vector
ret = odeint(deriv, y0, t, args=(N, beta))  # Integrate the SIR equations over the time grid, t.
S, I = ret.T
 
fig = plt.figure(figsize=(10,6))
ax = plt.axes(xlim=(0, days), ylim=(0, 1))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.plot(t, S/N, 'b', alpha=0.5, lw=2, label='Здоровые')
ax.plot(t, I/N, 'r', alpha=0.5, lw=2, label='Инфицированные')
ax.plot(t, (S+I)/N, 'k', alpha=0.5, lw=2, label='Сумма')
ax.set_xticks(np.arange(0, days, 30))
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.legend()
ax.set_xlabel('время / дни')
ax.set_ylabel('процент популяции')
plt.show()
```
