```python
from matplotlib import animation
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np
```

```python
def f(y,t):
  return y[1], -0.1*y[1]-np.sin(y[0]) #+ 2.321*np.exp(-10*(t-5)**4)

n, steps = 100, 1000
xs = np.linspace(0,n,steps)

y0 = [np.pi, 0]  # начальный угол отклонения, начальная угловая скорость
us = odeint(f,y0,xs)  # численное решение ДУ

plt.figure(figsize=(6,3), dpi = 100)
ys = us[:,0]          # найденные значения y(t) в точках, изменение угла с течением времени
plt.plot(xs,ys,'-')   # график для y(t)
plt.xlabel('время (секунды)')
plt.ylabel('y(t)')
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, 10)
plt.show()

plt.figure(figsize=(6,3), dpi = 100)
ds = us[:,1]          # найденные значения y'(t) в точках, угловая скорость с течением времени
plt.plot(xs,ds,'-')   # график для y'(t)
plt.xlabel('время (секунды)')
plt.ylabel('y\'(t)')
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(-2, 2)
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(xlim=(-1, 1), ylim=(-1, 1))
point, = plt.plot([], [], 'g.', ms=20, c='black')
line, = ax.plot([], [], '-', lw=2)
# ax.grid() # для сетки на графике
l = 0.9     # длина нити
plt.close(fig)
def animate(i):
    line.set_data([0,l*np.sin(ys[i])],[0,-l*np.cos(ys[i])])
    point.set_data(l*np.sin(ys[i]),-l*np.cos(ys[i]))
    return line,
anim = animation.FuncAnimation(fig, animate, frames=steps, interval=100, blit=True)
from IPython.display import HTML
HTML(anim.to_html5_video())
```
