```python
from sympy import*
t = symbols("t")
x = Function("x")
m = 1
k = 0.01
```

```python
de = Eq(m*diff(x(t),t,t),-k*x(t))
de
```

```python
s =  dsolve(de,x(t), ics={x(0):0.5, diff(x(t),t).subs(t,0):0})
s
```

```python
pt = plot(s.rhs, (t, 0,100))
```
